```python
import pandas as pd
df = pd.read_csv('../lichess_db_puzzle.csv')
df.head()
```

```python
df = df[['PuzzleId', 'FEN', 'Rating']]
df.head()
```

```python
# sort by Rating in ascending order
df = df.sort_values(by='Rating', ascending=True)
df.head()
```

```python
# select first 5000 rows
df = df.head(5000)

# I want to select random 500 rows for sft, 1000 rows for rl training, and 50 rows for evaluation
sft_df = df.sample(n=500, random_state=42)
remaining_df = df.drop(sft_df.index)
rl_df = remaining_df.sample(n=1000, random_state=42)
eval_df = remaining_df.drop(rl_df.index).sample(n=50, random_state=42)
sft_df.to_csv('sft_data.csv', index=False)
rl_df.to_csv('rl_data.csv', index=False)
eval_df.to_csv('eval_data.csv', index=False)
```
